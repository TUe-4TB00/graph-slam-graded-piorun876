```python
import sys
from pathlib import Path

env = Path(sys.executable).parents[1]
print("Python:", sys.executable)
print("Conda env:", env)

candidates = [
    env / "Library" / "bin" / "dot.exe",
    env / "bin" / "dot.exe",
]

for p in candidates:
    print(p, "exists:", p.exists())
```
